```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import mpld3

%matplotlib inline
```

```python
def parse_cross_csv(reader):
    time = []
    spectrum_data = []

    for row in reader:
        time.append(float(row[0]))
        spectrum = []
        for i in range(1,len(row)-1):
            spectrum.append(float(row[i]))
        spectrum_data.append(spectrum)

    time =  np.array(time)
    spectrum_data = np.array(spectrum_data)

    spectrum_data = np.transpose(spectrum_data)
    for spectrum in range(len(spectrum_data)):
        plt.plot(time, -1*spectrum_data[spectrum])


    plt.axis([0, 1, -15, 1])
    plt.title("Energy eigenvalues vs time")
    plt.xlabel("normalized time (s)")
    plt.ylabel("Energy relative to ground state")
    plt.show()
```

```python
xing1 = open('crossover/xover_1.csv', mode='r', encoding='utf-8-sig')
xing3 = open('crossover/xover_3.csv', mode='r', encoding='utf-8-sig')
xing1_reader = csv.reader(xing1)
xing3_reader = csv.reader(xing3)

parse_cross_csv(xing1_reader)
parse_cross_csv(xing3_reader)
```
